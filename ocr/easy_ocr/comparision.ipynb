```python
import matplotlib.pyplot as plt
import numpy as np

# Data extracted from the image
data = {
    "CPU": [1.94, 1.96, 1.94, 2.17, 2.8],
    "NPU": [4.92, 4.71, 5.18, 5.85, 12.61],
    "GPU": [2.9, 0.57, 0.47, 0.77, 0.92],
    "ONNX": [2.83, 2.07, 2.30, 3.35, 2.7]
}


# Creating a numpy array for x-axis
x = np.arange(1, len(data["CPU"]) + 1)

# Plotting the data
plt.figure(figsize=(10, 6))
for label, y in data.items():
    plt.plot(x, y, marker='o', label=label)

# Adding titles and labels
plt.title("Performance Comparison")
plt.xlabel("Inputs")
plt.ylabel("Time (s)")
plt.legend()
plt.grid(True)
plt.show()
```
